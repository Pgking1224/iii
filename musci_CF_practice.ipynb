```python
df = pd.merge(df,a,on='genre_id',how='inner')
```

```python
df = df[['id','genre_id','rank','rank_mean']]
```

```python
df['ad'] = df['rank'] - df['rank_mean']
df
```

```python
movie_data_all_append=pd.DataFrame()

user_data=  df[df['id']!='+06pc1V8QfywrGOTNln/7X7Md8dA8fVOIU37n2M5mv8']
distinct_movies=np.unique(user_data['genre_id'])
i=1
for movie in distinct_movies[10:11]:
    
    if i%10==0:
        
        print (i , "out of ", len(distinct_movies))
    
    movie_data_all=pd.DataFrame()
    
    movie_data = df[df['genre_id']==movie]
    movie_data = movie_data[['id','genre_id','ad']].drop_duplicates()
    movie_data=movie_data.rename(columns={'ad':'ad1'})
    movie_data=movie_data.rename(columns={'genre_id':'genre_id1'})
    movie1_val=np.sqrt(np.sum(np.square(movie_data['ad1']), axis=0))
    
    user_data1=  df[df['id']=='+06pc1V8QfywrGOTNln/7X7Md8dA8fVOIU37n2M5mv8']
    distinct_movies1=np.unique(user_data1['genre_id'])
    
    
    
    for movie1 in distinct_movies1:
        
        movie_data1 = df[df['genre_id']==movie1]
        movie_data1 = movie_data1[['id','genre_id','ad']].drop_duplicates()
        movie_data1=movie_data1.rename(columns={'ad':'ad2'})
        movie_data1=movie_data1.rename(columns={'genre_id':'genre_id2'})
        movie2_val=np.sqrt(np.sum(np.square(movie_data1['ad2']), axis=0))
        
        movie_data_merge = pd.merge(movie_data,movie_data1[['id','genre_id2','ad2']],on = 'id', how = 'inner', sort = False)
        
        movie_data_merge['vector_product']=(movie_data_merge['ad1']*movie_data_merge['ad2'])


        movie_data_merge= movie_data_merge.groupby(['genre_id1','genre_id2'], as_index = False, sort = False).sum()

        movie_data_merge['dot']=movie_data_merge['vector_product']/(movie1_val*movie2_val)
        
        movie_data_all = movie_data_all.append(movie_data_merge, ignore_index=True)
        
    
    movie_data_all=  movie_data_all[movie_data_all['dot']<1]
    movie_data_all = movie_data_all.sort_values(['dot'], ascending=False)
    movie_data_all = movie_data_all.head(20)
    
    movie_data_all_append = movie_data_all_append.append(movie_data_all, ignore_index=True)
    i=i+1
```

```python
movie_rating_all=pd.DataFrame()

for movie in distinct_movies[10:11]:
    movie_nbr=movie_data_all_append[movie_data_all_append['genre_id1']==movie]
    movie_mean = df[df['genre_id']==movie]
    mean = movie_mean["rank"].mean()
    movie_nbr_dot = pd.merge(user_data1,movie_nbr[['dot','genre_id2','genre_id1']], how = 'inner',left_on='genre_id', right_on='genre_id2', sort = False)
    movie_nbr_dot['wt_rating']=movie_nbr_dot['dot']*movie_nbr_dot['ad']
    movie_nbr_dot['dot_abs']=movie_nbr_dot['dot'].abs()
    movie_nbr_dot= movie_nbr_dot.groupby(['genre_id1'], as_index = False, sort = False).sum()[['genre_id1','wt_rating','dot_abs']]
    movie_nbr_dot['rating']=(movie_nbr_dot['wt_rating']/movie_nbr_dot['dot_abs'])+mean
    
    movie_rating_all = movie_rating_all.append(movie_nbr_dot, ignore_index=True)
    
movie_rating_all = movie_rating_all.sort_values(['rating'], ascending=False)
```
