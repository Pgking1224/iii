```python
id_list = list(df['msno'])
len(id_list)
```

```python
name_list = list(df['name'])
len(name_list)
```

```python
song_list = {}
for i in range(len(id_list)):
    if id_list[i] not in song_list:
        song_list[id_list[i]] = []
        song_list[id_list[i]].append(name_list[i])
    else:
        song_list[id_list[i]].append(name_list[i])
        
    print(id_list[i],name_list[i])
```
